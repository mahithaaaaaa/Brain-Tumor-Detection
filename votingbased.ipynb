```python
import torch
import torch.nn as nn
import torch.optim as optim
from torchvision import datasets, transforms, models
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, f1_score
from efficientnet_pytorch import EfficientNet
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from tqdm import tqdm
from google.colab import drive

# Mount Google Drive (for saving models)
drive.mount('/content/drive')

# Hyperparameters
batch_size = 32
learning_rate = 0.001
num_epochs = 5  # Set to 5 for brevity, you can increase it

# Dataset Paths
train_path = "/content/dataset/Testing"
test_path = "/content/dataset/Testing"

# Define Data Transformations
transform = transforms.Compose([
    transforms.Resize((224, 224)),  # Resize images
    transforms.ToTensor(),         # Convert to Tensor
    transforms.Normalize(mean=[0.5, 0.5, 0.5], std=[0.5, 0.5, 0.5])  # Normalize
])

# Load Datasets
train_dataset = datasets.ImageFolder(root=train_path, transform=transform)
test_dataset = datasets.ImageFolder(root=test_path, transform=transform)

train_loader = torch.utils.data.DataLoader(train_dataset, batch_size=batch_size, shuffle=True)
test_loader = torch.utils.data.DataLoader(test_dataset, batch_size=batch_size, shuffle=False)

# Device Configuration
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# Helper Function to Create Models
def create_model(model_name):
    if model_name == "resnet50":
        model = models.resnet50(pretrained=True)
        model.fc = nn.Linear(model.fc.in_features, len(train_dataset.classes))
    elif model_name == "efficientnet_b0":
        model = EfficientNet.from_pretrained("efficientnet-b0")
        model._fc = nn.Linear(model._fc.in_features, len(train_dataset.classes))
    elif model_name == "densenet":
        model = models.densenet121(pretrained=True)
        model.classifier = nn.Linear(model.classifier.in_features, len(train_dataset.classes))
    else:
        raise ValueError(f"Unsupported model name {model_name}")
    return model.to(device)

# Models to Train
models_to_train = ["resnet50", "efficientnet_b0", "densenet"]

# Train Function
def train_model(model, train_loader, criterion, optimizer):
    model.train()
    running_loss = 0
    correct = 0
    total = 0
    for images, labels in tqdm(train_loader):
        images, labels = images.to(device), labels.to(device)

        # Forward Pass
        outputs = model(images)
        loss = criterion(outputs, labels)

        # Backward Pass
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

        # Statistics
        running_loss += loss.item()
        _, predicted = torch.max(outputs, 1)
        total += labels.size(0)
        correct += (predicted == labels).sum().item()

    accuracy = 100 * correct / total
    return running_loss / len(train_loader), accuracy

# Test Function
def test_model(model, test_loader):
    model.eval()
    all_predictions = []
    all_labels = []
    with torch.no_grad():
        for images, labels in test_loader:
            images, labels = images.to(device), labels.to(device)

            outputs = model(images)
            _, predicted = torch.max(outputs, 1)

            all_predictions.extend(predicted.cpu().numpy())
            all_labels.extend(labels.cpu().numpy())

    return all_labels, all_predictions

# Confusion Matrix Visualization
def plot_confusion_matrix(cm, classes, title):
    plt.figure(figsize=(8, 6))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=classes, yticklabels=classes)
    plt.title(f"Confusion Matrix - {title}")
    plt.xlabel("Predicted Labels")
    plt.ylabel("True Labels")
    plt.show()

# Voting-Based Fusion Function
def voting_based_fusion(models, test_loader, weights=None):
    all_fused_preds = []
    all_labels = []

    for images, labels in tqdm(test_loader):
        images = images.to(device)
        batch_predictions = []

        # Collect predictions from each model
        for model_name, model in models.items():
            model.eval()
            with torch.no_grad():
                outputs = model(images)
                _, predicted = torch.max(outputs, 1)
                batch_predictions.append(predicted.cpu().numpy())

        # Perform voting
        batch_predictions = np.array(batch_predictions)  # Shape: (num_models, batch_size)
        if weights:
            weighted_votes = np.zeros((batch_predictions.shape[1], len(train_dataset.classes)))
            for i, model_name in enumerate(models.keys()):
                for j, pred_class in enumerate(batch_predictions[i]):
                    weighted_votes[j, pred_class] += weights[model_name]
            fused_preds = np.argmax(weighted_votes, axis=1)
        else:
            fused_preds = np.apply_along_axis(
                lambda x: np.bincount(x, minlength=len(train_dataset.classes)).argmax(),
                axis=0,
                arr=batch_predictions
            )

        all_fused_preds.extend(fused_preds)
        all_labels.extend(labels.numpy())

    return all_labels, all_fused_preds

# Evaluate Voting-Based Fusion
def evaluate_voting_fusion(models, test_loader, test_accuracies):
    print("\nEvaluating Voting-Based Fusion...")

    # Determine Weights for Weighted Voting (based on test accuracy)
    weights = {model_name: acc for model_name, acc in test_accuracies.items()}
    print(f"Weights for models (based on accuracy): {weights}")

    # Majority Voting
    print("\nPerforming Majority Voting...")
    majority_labels, majority_predictions = voting_based_fusion(models, test_loader)

    # Majority Voting Results
    majority_acc = accuracy_score(majority_labels, majority_predictions)
    majority_prec = precision_score(majority_labels, majority_predictions, average='macro')
    majority_f1 = f1_score(majority_labels, majority_predictions, average='macro')
    majority_cm = confusion_matrix(majority_labels, majority_predictions)

    print(f"\nMajority Voting Results:\nAccuracy: {majority_acc:.2f}, Precision: {majority_prec:.2f}, F1 Score: {majority_f1:.2f}")
    plot_confusion_matrix(majority_cm, train_dataset.classes, "Majority Voting")

    # Weighted Voting
    print("\nPerforming Weighted Voting...")
    weighted_labels, weighted_predictions = voting_based_fusion(models, test_loader, weights=weights)

    # Weighted Voting Results
    weighted_acc = accuracy_score(weighted_labels, weighted_predictions)
    weighted_prec = precision_score(weighted_labels, weighted_predictions, average='macro')
    weighted_f1 = f1_score(weighted_labels, weighted_predictions, average='macro')
    weighted_cm = confusion_matrix(weighted_labels, weighted_predictions)

    print(f"\nWeighted Voting Results:\nAccuracy: {weighted_acc:.2f}, Precision: {weighted_prec:.2f}, F1 Score: {weighted_f1:.2f}")
    plot_confusion_matrix(weighted_cm, train_dataset.classes, "Weighted Voting")

# Training and Evaluating Models
trained_models = {}
test_accuracies = {}

for model_name in models_to_train:
    print(f"\nTraining {model_name}...")
    model = create_model(model_name)
    criterion = nn.CrossEntropyLoss()
    optimizer = optim.Adam(model.parameters(), lr=learning_rate)

    for epoch in range(num_epochs):
        train_loss, train_accuracy = train_model(model, train_loader, criterion, optimizer)
        print(f"Epoch [{epoch+1}/{num_epochs}] - Loss: {train_loss:.4f}, Accuracy: {train_accuracy:.2f}%")

    # Test the model
    labels, predictions = test_model(model, test_loader)
    acc = accuracy_score(labels, predictions)
    test_accuracies[model_name] = acc
    print(f"Test Accuracy for {model_name}: {acc:.2f}")

    # Save model
    model_save_path = f"/content/drive/MyDrive/{model_name}_brain_tumor.pth"
    torch.save(model.state_dict(), model_save_path)
    print(f"Model saved to: {model_save_path}")

    trained_models[model_name] = model

# Perform Voting-Based Fusion
evaluate_voting_fusion(trained_models, test_loader, test_accuracies)
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
from torchvision import datasets, transforms, models
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, f1_score
from efficientnet_pytorch import EfficientNet
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from tqdm import tqdm
from google.colab import drive

# Mount Google Drive (for saving models)
drive.mount('/content/drive')

# Hyperparameters
batch_size = 32
learning_rate = 0.001
num_epochs = 10  # Set the number of epochs for training

# Dataset Paths
train_path = "/content/dataset/Training"  # Replace with your training dataset path
test_path = "/content/dataset/Testing"    # Replace with your testing dataset path

# Define Data Transformations
transform = transforms.Compose([
    transforms.Resize((224, 224)),  # Resize images
    transforms.ToTensor(),         # Convert to Tensor
    transforms.Normalize(mean=[0.5, 0.5, 0.5], std=[0.5, 0.5, 0.5])  # Normalize
])

# Load Datasets
train_dataset = datasets.ImageFolder(root=train_path, transform=transform)
test_dataset = datasets.ImageFolder(root=test_path, transform=transform)

train_loader = torch.utils.data.DataLoader(train_dataset, batch_size=batch_size, shuffle=True)
test_loader = torch.utils.data.DataLoader(test_dataset, batch_size=batch_size, shuffle=False)

# Device Configuration
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# Create Models
def create_model(model_name):
    if model_name == "resnet50":
        model = models.resnet50(pretrained=True)
        model.fc = nn.Linear(model.fc.in_features, len(train_dataset.classes))
    elif model_name == "efficientnet_b0":
        model = EfficientNet.from_pretrained("efficientnet-b0")
        model._fc = nn.Linear(model._fc.in_features, len(train_dataset.classes))
    elif model_name == "densenet":
        model = models.densenet121(pretrained=True)
        model.classifier = nn.Linear(model.classifier.in_features, len(train_dataset.classes))
    return model.to(device)

# Models to Train
models_to_train = ["resnet50", "efficientnet_b0", "densenet"]

# Train Function with Tracking of Loss and Accuracy
def train_model(model, train_loader, criterion, optimizer, num_epochs):
    model.train()
    epoch_losses = []
    epoch_accuracies = []

    for epoch in range(num_epochs):
        running_loss = 0
        correct = 0
        total = 0

        for images, labels in tqdm(train_loader):
            images, labels = images.to(device), labels.to(device)

            # Forward Pass
            outputs = model(images)
            loss = criterion(outputs, labels)

            # Backward Pass
            optimizer.zero_grad()
            loss.backward()
            optimizer.step()

            # Statistics
            running_loss += loss.item()
            _, predicted = torch.max(outputs, 1)
            total += labels.size(0)
            correct += (predicted == labels).sum().item()

        # Compute Loss and Accuracy for the current epoch
        epoch_loss = running_loss / len(train_loader)
        epoch_accuracy = 100 * correct / total
        epoch_losses.append(epoch_loss)
        epoch_accuracies.append(epoch_accuracy)

        print(f"Epoch [{epoch+1}/{num_epochs}] - Loss: {epoch_loss:.4f}, Accuracy: {epoch_accuracy:.2f}%")

    return model, epoch_losses, epoch_accuracies

# Testing Function
def test_model(model, test_loader):
    model.eval()
    all_predictions = []
    all_labels = []

    with torch.no_grad():
        for images, labels in test_loader:
            images, labels = images.to(device), labels.to(device)

            outputs = model(images)
            _, predicted = torch.max(outputs, 1)

            all_predictions.extend(predicted.cpu().numpy())
            all_labels.extend(labels.cpu().numpy())

    return all_labels, all_predictions

# Plot Loss vs Epochs and Accuracy vs Epochs
def plot_training_curves(epoch_losses, epoch_accuracies, model_name):
    # Plot Loss vs Epochs
    plt.figure(figsize=(10, 5))
    plt.plot(range(1, len(epoch_losses) + 1), epoch_losses, label="Loss", color="red")
    plt.title(f"Loss vs Epochs for {model_name}")
    plt.xlabel("Epochs")
    plt.ylabel("Loss")
    plt.legend()
    plt.grid()
    plt.show()

    # Plot Accuracy vs Epochs
    plt.figure(figsize=(10, 5))
    plt.plot(range(1, len(epoch_accuracies) + 1), epoch_accuracies, label="Accuracy", color="blue")
    plt.title(f"Accuracy vs Epochs for {model_name}")
    plt.xlabel("Epochs")
    plt.ylabel("Accuracy (%)")
    plt.legend()
    plt.grid()
    plt.show()

# Confusion Matrix Visualization
def plot_confusion_matrix(cm, classes, title):
    plt.figure(figsize=(8, 6))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=classes, yticklabels=classes)
    plt.title(f"Confusion Matrix - {title}")
    plt.xlabel("Predicted Labels")
    plt.ylabel("True Labels")
    plt.show()

# Voting-Based Fusion Function
def voting_based_fusion(models, test_loader, weights=None):
    all_fused_preds = []
    all_labels = []

    for images, labels in tqdm(test_loader):
        images = images.to(device)
        batch_predictions = []

        # Collect predictions from each model
        for model_name, model in models.items():
            model.eval()
            with torch.no_grad():
                outputs = model(images)
                _, predicted = torch.max(outputs, 1)
                batch_predictions.append(predicted.cpu().numpy())

        # Perform voting
        batch_predictions = np.array(batch_predictions)
        if weights:
            weighted_votes = np.zeros((batch_predictions.shape[1], len(train_dataset.classes)))
            for i, model_name in enumerate(models.keys()):
                for j, pred_class in enumerate(batch_predictions[i]):
                    weighted_votes[j, pred_class] += weights[model_name]
            fused_preds = np.argmax(weighted_votes, axis=1)
        else:
            fused_preds = np.apply_along_axis(
                lambda x: np.bincount(x, minlength=len(train_dataset.classes)).argmax(),
                axis=0,
                arr=batch_predictions
            )

        all_fused_preds.extend(fused_preds)
        all_labels.extend(labels.numpy())

    return all_labels, all_fused_preds

# Evaluate Voting-Based Fusion
def evaluate_voting_fusion(models, test_loader, test_accuracies):
    print("\nEvaluating Voting-Based Fusion...")

    # Majority Voting
    print("\nPerforming Majority Voting...")
    majority_labels, majority_predictions = voting_based_fusion(models, test_loader)

    # Majority Voting Confusion Matrix and Results
    majority_cm = confusion_matrix(majority_labels, majority_predictions)
    majority_acc = accuracy_score(majority_labels, majority_predictions)
    majority_prec = precision_score(majority_labels, majority_predictions, average='macro')
    majority_f1 = f1_score(majority_labels, majority_predictions, average='macro')
    print(f"Majority Voting Results:\nAccuracy: {majority_acc:.2f}, Precision: {majority_prec:.2f}, F1 Score: {majority_f1:.2f}")
    plot_confusion_matrix(majority_cm, train_dataset.classes, "Majority Voting")

# Training and Plotting for Each Model
trained_models = {}
test_accuracies = {}

for model_name in models_to_train:
    print(f"\nTraining {model_name}...")
    model = create_model(model_name)
    criterion = nn.CrossEntropyLoss()
    optimizer = optim.Adam(model.parameters(), lr=learning_rate)

    # Train the model
    model, epoch_losses, epoch_accuracies = train_model(model, train_loader, criterion, optimizer, num_epochs)

    # Plot Training Curves
    plot_training_curves(epoch_losses, epoch_accuracies, model_name)

    # Test the model
    labels, predictions = test_model(model, test_loader)
    acc = accuracy_score(labels, predictions)
    test_accuracies[model_name] = acc
    print(f"Test Accuracy for {model_name}: {acc:.2f}")

    # Save model
    model_save_path = f"/content/drive/MyDrive/votedFusion_brain_tumor.pth"
    torch.save(model.state_dict(), model_save_path)
    print(f"Model saved to: {model_save_path}")

    trained_models[model_name] = model

# Perform Voting-Based Fusion
evaluate_voting_fusion(trained_models, test_loader, test_accuracies)
```
